## How to convert a zip code into latitude and longitude values

```python
# Install the pgeocode library
!pip install pgeocode

# Example of postal code query
import pgeocode

# Create an instance for United States ('us')
nomi = pgeocode.Nominatim('us')

# Check zip code 10024 (New York City)
result = nomi.query_postal_code("10024")

# Get the latitude and longitude coordinates
latitud = result['latitude']
longitud = result['longitude']

print(latitud)
print(longitud)

print('Lets visit ! (https://maps.app.goo.gl/ozRQ8QBuN18MRxP36)')
```
